```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../2.DATA/daily_incom_per_capita.csv")
```

```python
print(df.iloc[:, 1:].dtypes)  # Ensure data types are float
print(df.head())  # Check displayed values
```

```python
df.iloc[:, 1:] = (df.iloc[:, 1:].astype(float) * 365.5).round(2)

updated_csv_path = "../2.DATA/annual_income_per_capita_updated.csv"

df.to_csv(updated_csv_path, index=False, float_format="%.2f")

# Provide the downloadable link
updated_csv_path
```
